```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf


data = pd.read_csv("base_modelo_supervisado.csv")

data.head()
```

```python
data.describe()
```

```python
data.corr()
```

```python
model = smf.ols('Y ~ X1 + X2 + X3 + X4 + X5 + X6 + X7 + X8', data = data)
results = model.fit()
print(results.summary())
```

```python

# Coeficiente de determinación R^2
r_squared = results.rsquared
print("Coeficiente de determinación R^2:", r_squared)
```

```python
datos = pd.read_csv("base_modelo_nosupervisado.csv")

datos.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler



kmeans = KMeans(n_clusters=3)  # Define el número de clusters que deseas encontrar
kmeans.fit(datos[['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8', 'X9', 'X10']])

datos['Cluster'] = kmeans.labels_

datos.head(20)
```

```python
import matplotlib.pyplot as plt
inertia = []
for n_clusters in range(1, 11):  # Prueba con diferentes números de clusters de 1 a 10
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(datos[['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8', 'X9', 'X10']])
    inertia.append(kmeans.inertia_)

# Graficar la inercia en función del número de clusters
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Gráfico de Inercia')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.show()
```

```python
cluster_counts = datos['Cluster'].value_counts()

# Mostrar el número de observaciones en cada cluster
print(cluster_counts)
```

```python
for cluster_num in range(3):  # Iterar sobre los números de cluster 0, 1, 2
    count_in_cluster = len(datos[datos['Cluster'] == cluster_num]['X1'])
    print(f"Cantidad de elementos de X1 en el cluster {cluster_num}: {count_in_cluster}")
```

```python
for cluster_num in range(3):  # Iterar sobre los números de cluster 0, 1, 2
    count_in_cluster = len(datos[datos['Cluster'] == cluster_num]['X2'])
    print(f"Cantidad de elementos de X1 en el cluster {cluster_num}: {count_in_cluster}")
```

```python
# Encontrar el cluster con la mayor cantidad de observaciones
cluster_mas_grande = cluster_counts.idxmax()

# Obtener una observación dentro del cluster más grande
observacion_en_cluster_mas_grande = datos[datos['Cluster'] == cluster_mas_grande].iloc[0]

# Obtener el valor de X1 para esta observación
valor_X1_cluster_mas_grande = observacion_en_cluster_mas_grande['X1']

print("Valor de X1 en el segmento más grande:", valor_X1_cluster_mas_grande)
```

```python
# Calcular la varianza de cada variable dentro de cada cluster
varianza_por_variable = datos.groupby('Cluster').var()

# Calcular la varianza total de cada variable sumando las varianzas de los clusters
varianza_total_por_variable = varianza_por_variable.sum()

# Ordenar las variables por su varianza total en orden descendente
variables_ordenadas = varianza_total_por_variable.sort_values(ascending=False)

# Imprimir el orden de importancia discriminante de las variables
print("Orden de importancia discriminante de las variables:")
print(variables_ordenadas)
```
