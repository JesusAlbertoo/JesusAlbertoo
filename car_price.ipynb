```python
import pandas as pd
datos = pd.read_csv("cardekho.csv")

datos = datos.drop("seats", axis=1)
datos.head()
```

```python
datos.info()
```

```python
datos["max_power"].dropna()
```

```python
datos['max_power'] = pd.to_numeric(datos['max_power'], errors='coerce')
datos.info()
```

```python
mileage_mean = datos["mileage(km/ltr/kg)"].mean()
engine_mean = datos["engine"].mean()
max_power_mean = datos["max_power"].mean()
```

```python
datos["mileage(km/ltr/kg)"] = datos["mileage(km/ltr/kg)"].fillna(mileage_mean)
datos["engine"] = datos["engine"].fillna(engine_mean)
datos["max_power"] = datos["max_power"].fillna(max_power_mean)
```

```python
import numpy as np
import plotly.express as px

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import preprocessing



from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing

#Escalamiento
scale = StandardScaler()

datos_escalados = scale.fit_transform(datos[["year", "km_driven",  "mileage(km/ltr/kg)", "engine", "max_power"]])

dato_a_predecir = scale.fit_transform(datos[["selling_price"]])

#separamos el dataset en los datos que entrenaremos y los queocuparemos para probar los modelos

X_train, X_test, y_train, y_test = train_test_split(datos_escalados, dato_a_predecir, random_state=42, test_size=0.2)
```

```python
#regresin lineal
from sklearn.metrics import mean_squared_error
regresion_lineal_model = LinearRegression().fit(X_train, y_train)
prediccion_lineal = regresion_lineal_model.predict(X_test)
mse = mean_squared_error(y_test, prediccion_lineal)

print('Coeficientes:', regresion_lineal_model.coef_)
print('Trmino independiente:', regresion_lineal_model.intercept_)

print("El error es:", mse)
```

```python
#Vamos a hacer la regresin con statsmodel

import statsmodels.api as sm

X = sm.add_constant(X_train)

# Ajustar el modelo de regresin lineal mltiple
model = sm.OLS(y_train, X).fit()

# Imprimir los resultados del modelo
print(model.summary())
```
