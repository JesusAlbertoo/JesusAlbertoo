```python
from numpy import *
from scipy.integrate import *
from matplotlib.pyplot import *
```

```python
def sir(y, T, N, beta, gama):
    S, I, R = y
    dSdt = -beta*((S*I))
    dIdt = beta*((S*I)) - gama*I
    dRdt = gama*I
    return dSdt, dIdt, dRdt
```

```python
def plotSIR(S, I, R, t, normaliza=1):
    fig, ax = subplots()
    ax.plot(t, S/normaliza,'b', alpha = 0.5, lw=2, label='Susceptibles')
    ax.plot(t, I/normaliza,'r', alpha = 0.5, lw=2, label='Infecciosos')
    ax.plot(t, R/normaliza,'g', alpha = 0.5, lw=2, label='Removidos')
    ax.set_xlabel('tiempo')
    ax.set_ylabel(r'Poblaci√≥n normalizada  (%d)' %normaliza)
    legend = ax.legend()
    show()
```

```python
N=261
I0=7
R0=0
S0=N-I0-R0
beta=0.0173653
gama=2.76518
t=linspace(0,5,100)
y0 = S0, I0, R0
```

```python
sol=odeint(sir, y0, t, args=(N, beta, gama))
```

```python
S, I, R = sol.T
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def system(variables, t, beta, alpha, phi):
    S, Z, R = variables
    dSdt = -beta * S * Z
    dZdt = beta * S * Z + phi * R - alpha * S * Z
    dRdt = alpha * S * Z - phi * R
    return [dSdt, dZdt, dRdt]

# Par√°metros
beta = 0.1
alpha = 0.9
phi = 0.5

# Condiciones iniciales
S0 = 2
Z0 = 0.3
R0 = 0.0

# Tiempo
t = np.linspace(0, 100, 100)  # Intervalo de tiempo 

# Soluci√≥n de las ecuaciones diferenciales
sol = odeint(system, [S0, Z0, R0], t, args=(beta, alpha, phi))

# Gr√°fica
plt.plot(t, sol[:, 0], label='S')
plt.plot(t, sol[:, 1], label='Z')
plt.plot(t, sol[:, 2], label='R')
plt.xlabel('Tiempo')
plt.ylabel('Cantidad')
plt.title('Sistema de Ecuaciones Diferenciales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir las ecuaciones diferenciales del modelo Lotka-Volterra
def modelo(y, t):
    x, y1, y2 = y
    dxdt = alpha*x
    dy1dt = gama1*y1 - delta1*y1*x
    dy2dt = gama2*y2 - delta2*y2*x
    return [dxdt, dy1dt, dy2dt]

# Definir los par√°metros del modelo
alpha = 0.00378
beta1 = 0.0
beta2 = 0.00
gama1 = 0.041
gama2 = 0.02
delta1 = 0.075
delta2 = 0.015

# Condiciones iniciales
condiciones_iniciales = [2, 20, 100]

# Definir el tiempo de integraci√≥n
t = np.linspace(0, 200, 1000)

# Resolver las ecuaciones diferenciales
solution = odeint(modelo, condiciones_iniciales, t)

# Graficar las soluciones
plt.plot(t, solution[:, 0], label='Depredador (x)')
plt.plot(t, solution[:, 1], label='Presa 1 (y1)')
plt.plot(t, solution[:, 2], label='Presa 2 (y2)')
plt.xlabel('Tiempo')
plt.ylabel('Poblaci√≥n')
plt.title('Modelo Lotka-Volterra')
plt.legend()
plt.show()
```
