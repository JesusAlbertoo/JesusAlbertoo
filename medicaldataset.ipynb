```python
#Vamos a cambiar las etiquetas en Results
import numpy as np
import plotly.express as px

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

LabelEncoder()
datos["Result"] = LabelEncoder().fit_transform(datos["Result"])
```

```python
#vamos a escarlar los valores
X = datos.iloc[:, :-1].values
y = datos.iloc[: ,-1].values



scaler = MinMaxScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
X
```

```python
def training(model, name, axes):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    score = accuracy_score(y_test, pred)
    report = classification_report(y_test, pred)
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, ax=axes)
    axes.set_title("{}: {}%".format(name, round(score*100, 2)))
    return score*100, report
```

```python
rfc = RandomForestClassifier()
etc = ExtraTreesClassifier()
gbc = GradientBoostingClassifier()
abc = AdaBoostClassifier()
lgr = LogisticRegression()
xgb = XGBClassifier()


models = [rfc, etc, gbc, abc, lgr, xgb]
names = ["Random Forest", "Extra Trees", "Gradient Boosting",
        "Ada Boost", "Logistic Regression", "XGBoost"]
```

```python
reports, scores = [], []
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 7))
index = 0
for i in range(2):
    for j in range(3):
        score, report = training(models[index], names[index], axes[i][j])
        scores += [score]
        reports += [report]
        index += 1
plt.tight_layout()
plt.show()
```

```python
dt = pd.DataFrame({"scores": scores}, index=names)
dt = dt.sort_values("scores", ascending=False)
dt["scores"] = round(dt["scores"], 2)
fig, axes = plt.subplots()
sns.barplot(x=dt.index, y=dt.iloc[:, 0], ax=axes)
for container in axes.containers:
    axes.bar_label(container)
axes.set_xticklabels(axes.get_xticklabels(), rotation=90)
plt.show()
```

```python

for i, j in zip(names, reports):
    print("*******"*10)
    print(i)
    print(j)
    print("\n\n")
```
