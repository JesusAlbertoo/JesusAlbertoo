```python
#almacenamos la información de los datos
M = np.array([[1, 2.31],[2, 2.01], [3, 1.80], [4, 1.66], [5, 1.55],[6, 1.47], [7, 1.41]])

#construiremos la matriz V
V = np.array([ [ 1.0, M[i,0], M[i,0]**2] for i in range(len(M))])
print('La matriz V es:\n',V)

#construiremos el vector b
b = np.array([ M[i,1] for i in range(len(M))])
print('\nEl vector b es:\n',b)

#Resolveremos el sistema de ecuaciones normales.
MV = np.dot(V.T,V)   #Producto de V^T por V
Mb = np.dot(V.T,b)   #Producto de V^T por b
a = np.linalg.solve(MV,Mb) 

print('\nEl vector de coeficientes de P(x) es:\n',a)

def P(x):
    y = a[0] + a[1]*x + a[2]*x**2    
    return y

#gráfica de P(x)
X = linspace(1,12)
Y = np.array([P(X[i]) for i in range(len(X))])
plot(X,Y,label='P(x)')

#Volveremos a usar a la matriz M para graficar a los puntos de S
plot(M[:,0],M[:,1],'*',color='r',label='Puntos de S')
xlabel(r'$x$')
ylabel(r'$y$')
grid()
legend()
show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Puntos dados
points = np.array([[1.02, 0.39], [0.95, 0.32], [0.87, 0.27], [0.77, 0.22], [0.67, 0.18], 
                   [0.56, 0.15], [0.44, 0.13], [0.30, 0.12], [0.16, 0.13], [0.01, 0.15]])

# Construir matriz A y vector b
x = points[:, 0]
y = points[:, 1]
A = np.vstack([x**2, x*y, y**2, x, y, np.ones(len(points))]).T
b = np.ones(len(points)) * -1

# Resolver el sistema de ecuaciones para obtener a, b, h y k
solution = np.linalg.lstsq(A, b, rcond=None)[0]
a2, ab, b2, ah, bk, c = solution
a = np.sqrt(a2)
b = np.sqrt(b2)
h = -ah / a2
k = -bk / b2

# Graficar la elipse con los puntos dados
theta = np.linspace(0, 2*np.pi, 100)
ellipse_x = h + a*np.cos(theta)
ellipse_y = k + b*np.sin(theta)
plt.scatter(points[:, 0], points[:, 1])
plt.plot(ellipse_x, ellipse_y)
plt.show()
```

```python
#Ejercicio 3

import numpy as np
import matplotlib.pyplot as plt

# Datos de la tabla
data = np.array([[-1.50, -0.997495],
                 [-1.25, -0.948985],
                 [-1.00, -0.841471],
                 [-0.75, -0.681639],
                 [-0.50, -0.479426],
                 [-0.250, -0.247404],
                 [0.00, 0.0],
                 [0.250, 0.247404],
                 [0.50, 0.479426],
                 [0.75, 0.681639],
                 [1.00, 0.841471],
                 [1.25, 0.948985],
                 [1.50, 0.997495]])

x = data[:, 0]
y = data[:, 1]

# Grados de polinomios de ajuste
degrees = [1, 3, 5, 7, 9, 11]

for n in degrees:
    # Construir matriz A y vector b
    A = np.vander(x, n + 1, increasing=True)
    b = y.reshape(len(y), 1)

    # Resolver el sistema de ecuaciones lineales para obtener los coeficientes
    coef = np.linalg.solve(A.T @ A, A.T @ b)

    # Graficar el polinomio de ajuste junto con los datos de la tabla
    xp = np.linspace(x.min(), x.max(), 100)
    yp = np.polyval(coef[::-1], xp)
    plt.plot(xp, yp, label=f'Polinomio de grado {n}')
    plt.scatter(x, y)

    # Calcular los errores absolutos de ajuste para todos los datos
    errors = np.abs(np.polyval(coef[::-1], x) - y)
    print(f'Error absoluto de ajuste para polinomio de grado {n}: {errors}')

plt.legend()
plt.show()
```

```python
#Ejercicio 4
from math import*
import numpy as np

print('Este programa calcula la factorización QR de una matriz A')
A = np.array([[1.0,1.0,1.5],[1.5,1.0,1.0],[1.0,1.5,1.0],[0.0,0.5,-0.5],[-0.5,0.5,0.0]])
B = np.copy(A) #Haremos una copia de A para usarla hasta el final

print('\nLa matriz A es:')
print(A)

#Función que construye a los vectores de Householder
def V(x):
    d = len(x)
    e1 = np.array([[0.0] for i in range(d)])
    e1[0,0] = 1.0
    if x[0,0] >= 0:
        alpha = -np.linalg.norm(x)
    else:
        alpha = np.linalg.norm(x)
    v = x - alpha*e1
    v = v/v[0,0]
    return v

#Función que construye a la matriz de Householder
def H(x):
    d = len(x)
    beta = 2.0/((np.linalg.norm(x))**2)
    H = np.identity(d) - beta*np.dot(x,x.T)
    return H

#Cálculo de la matriz Q y R
U = np.array([[0.0 for j in range(5)] for i in range(5)])
QT = np.identity(5)
for j in range(3):
    k = 5 - j
    a = np.array([[0.0] for i in range(k)])
    for i in range(k):
        a[i,0] = A[i+j,j]
    v = V(a)
    H1 = H(v)

    if j == 0:
        U = H1
    else:
        for i in range(5):
            for q in range(5):
                if i>=q:
                    U[q,q] = 1.0
        for i in range(j,5):
            for q in range(j,5):
                U[i,q] = H1[i-j,q-j]

    A = np.dot(U,A)           #La matriz R se almacenará en A
    QT = np.dot(U,QT)         #Matriz transpuesta de Q
    U = np.array([[0.0 for j in range(5)] for i in range(5)])

print('*******************************************')
print('\nLa matriz R es:')
print(A)                      #La matriz R se almacenó en A
print('\nLa matriz Q es')
print(QT.T)
print('\n*******************************************')
print('\nEl producto QR es')
print(np.dot(QT.T,A))
print('\n*******************************************')
print(np.dot(QT.T,A)==B)    #¿QR es igual a A? (Recuerda que B es una copia de la matriz original A)
```
